```python
import folium 
import pandas as pd
```

To create a folium map, we ned to provide the latitude and longitude  
Latitude -- north-south position (if south of equator, use a negative)  
Longitude -- east-west position (if west of the Greenwich mean lina, use a negative)  


```python
folium.Map(location=[40.7484, -73.9857])
```

```python
m=folium.Map(location=[40.7484, -73.9857])
m
```

```python
#Lets add a marker to the Empire State Building

folium.Marker([40.7484, -73.9857], popup='Empire State Building', tooltip='Click me!').add_to(m)
m
```

```python
p=folium.Map(location=[40.8200, -73.9493])
folium.Marker([40.8200, -73.9857], popup="City College of New York", tooltip='Click for more info', icon=folium.Icon(color='red')).add_to(p)
p
```

Make a new folium map, called `bins_map`.  
Center the map at 40.7128N, 74.006W (these are NYC coodinates)  

```python
bins_map=folium.Map(location=[40.7128, -74.006])
bins_map
```

```python
for i in range(10):
    row=bins.iloc[i]
    print("Coordinates ", row['Latitude'], row['Longitude'])
```

```python
for i in range(bins.shape[0]):
    row=bins.iloc[i]
    folium.Marker([row['Latitude'], row['Longitude']],
                  tooltip=row["Park/Site Name"]).add_to(bins_map)

bins_map
```
