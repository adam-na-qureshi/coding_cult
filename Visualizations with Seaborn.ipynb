```python
import seaborn as sns
%matplotlib inline
```

# Distribution Plots

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips
```

```python
sns.distplot(tips['total_bill']) #kde, kernal density estimate
```

```python
sns.distplot(tips['total_bill'], kde=False, rug=True)
```

```python
sns.distplot(tips['total_bill'], bins=35, kde=False, hist_kws=dict(edgecolor="black", linewidth=2))
plt.xlabel('Total Bill')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.ylabel('The Tip Amount')
plt.xlabel('The Total Bill', x=-3)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', color='purple')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm', kind='reg', diag_kind='hist')
```

# Categorical Plots

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

# Boxplot

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

# Stripplot()

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```
