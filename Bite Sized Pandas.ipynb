```python
flights.columns
```

```python
#Selecing a column

flights['ORIGIN']
```

```python
#Select multiple coluimns

flights[['ORIGIN', 'DEST']]
```

```python
#Select the first few rows
flights[:3]
```

```python
#Select a row and column entry in the DataFrame
flights.iloc[0, 0]
#iloc stands for integer locater 
```

```python
#Selecting the third row and the second column
flights.iloc[2, 1]
```

```python
#Mixing row indices with string column names
flights.iloc[2, flights.columns.get_loc('DAY_OF_MONTH')]
```

```python
#Sort values by a column least to greatest
flights.sort_values(by=['DISTANCE'])
```

```python
#Sort values by a column greatest to least
flights.sort_values(by=['DISTANCE'], ascending=False)
```

```python
#Sort values by multiple columns
flights.sort_values(by=['DISTANCE', 'AIR_TIME'], ascending=False)
```
